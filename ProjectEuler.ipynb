```python
'''
Problem 1

Output: Sum of all multiples of 3 or 5 to below 1000
'''
def multiples():
    total = 0
    for number in range(1, 1000):
        if number % 3 == 0 or number % 5 == 0:
            total += number
    return total

multiples()
    
```

```python
'''
Problem 2

Output: Sum of even values of Fibonacci terms/ max value cannot 
        exceed 4 million
'''
def fibonacci_sequence_total_to_four_million():
    fib = [1, 2]
    fib_seq = True
    total = 0
    while fib_seq:
        fib.append(fib[-1] + fib[-2])
        for i in fib:
            if i > 4000000:
                fib_seq = False
    for number in fib:
        if number % 2 == 0:
            total += number
    return total

fibonacci_sequence_total_to_four_million()
```

```python
'''
Problem 3

Output: Largest prime factor of 600851475143
'''
def is_prime_number(numb):
    for x in range(2, numb):
        if numb % x == 0:
            return False
    return True

def biggest_prime_factor():
    huge_number = 600851475143
    for i in range(2, huge_number):
        if huge_number % i == 0:
            maybe_prime_number = huge_number // i
            if is_prime_number(maybe_prime_number):
                return maybe_prime_number

biggest_prime_factor()
```

```python
'''
Problem 4
Output: Largest palindrome from the product of two 3-digit numbers
'''

def is_palindrome(numb):
    palindrome_check = str(numb)
    if palindrome_check == palindrome_check[::-1]:
        return True
    else:
        return False
    
def find_three_digit_palindrome():
    palindromes = []
    for first_numb in range(999, -1, -1):
        for second_numb in range(999, -1, -1):
            if is_palindrome(first_numb * second_numb):
                palindromes.append(first_numb * second_numb)
    return max(palindromes)

find_three_digit_palindrome()
```

```python
'''
Problem 5
Output: Smallest pos number evenly divisible by all numbers 1-20
'''

def smallest_multiple():
    guess = 2500
    while True:
        guess += 2
        for divisor in range(2, 21):
            if guess % divisor != 0:
                break
        else:
            return guess
        
smallest_multiple()
    
```

```python
'''
Problem 6
Output: Difference between sum of the squares of the first 100
        natural numbers and the square of the sum
'''

def square_sum_difference():
    total_addition_then_squared = 0
    total_addition_after_squared = 0
    for x in range(1, 101):
        total_addition_after_squared += x**2
        total_addition_then_squared += x
    return total_addition_then_squared**2 - total_addition_after_squared

square_sum_difference()
```

```python
'''
Problem 9
Output: product abc of pythagorean triplet where a + b + c = 1000
'''
import math

def pythagorean_triplet():
    for b in range(1, 1000):
        for a in range(0, b):
            c = math.sqrt(a**2 + b**2)
            if c - int(c) == 0.0:
                if a + b + c == 1000:
                    return a*b*c
            
print(pythagorean_triplet())
        
```

```python
'''
Problem 12
Output: Value of first triangle number > 500 divisors
'''
# 1 Add natural number and count position
# 2 Find factors and count them
# 3 If number of factors < 500 keep going
import time

def triangle_number():
    natural_number_count = 1
    factor_count = 0
    natural_number = 0
    even_factors = {}
    
    while factor_count < 500:
        natural_number += natural_number_count
        if natural_number % 2 != 0:
            for factor in range(1, natural_number + 1, 2):
                if natural_number % factor == 0:
                    factor_count += 1
        else:
            if natural_number // 2 in even_factors.keys():
                factor_count = len(even_factors[natural_number // 2]) + 1
                even_factors[natural_number] = even_factors[natural_number // 2] + [natural_number]
                even_factors.pop(natural_number // 2)
            else:
                for factor in range(1, natural_number + 1):
                    if natural_number % factor == 0:
                        try:
                            even_factors[natural_number].append(factor)
                            factor_count += 1
                        except KeyError:
                            even_factors[natural_number] = [factor]
                            factor_count += 1
                            
        if factor_count < 500:
            factor_count = 0
            natural_number_count += 1
    
    return natural_number

start_time = time.time()
print(triangle_number())
print(time.time() - start_time)

```

```python
'''
Problem 14
Output: starting number under 1 mil produces longest chain in colletz
'''
# 1. start at 1 million - 1, only checking odd numbers
# 2. perform operation, increase chain count
# 3. go until you get to 1,
import time

def longest_collatz_seq():
    chain_count = 1
    chain_count_max = 1
    huge_number = 999999
    tries = 0
    real_tries = 0
    huge_number_test = huge_number
    answer = 0
    while tries < 999999:
        if huge_number_test % 2 == 0:
            huge_number_test //= 2
            chain_count += 1
        else:
            huge_number_test = (huge_number_test * 3) + 1
            chain_count += 1
        if huge_number_test == 1:
            if chain_count > chain_count_max:
                chain_count_max = chain_count
                answer = huge_number
                real_tries += 1
            chain_count = 1
            tries += 1
            huge_number -= 1
            huge_number_test = huge_number
            
    return chain_count_max, real_tries, answer

start_time = time.time()
print(longest_collatz_seq())
print(time.time() - start_time)
```

```python
'''
Problem 17
Output: # letters in 1-1000 if written out
'''
import time, re

written_numbers = {}
written_numbers['1'] = ['one']
written_numbers['2'] = ['two', 'twenty']
written_numbers['3'] = ['three', 'thirty']
written_numbers['4'] = ['four', 'forty']
written_numbers['5'] = ['five', 'fifty']
written_numbers['6'] = ['six', 'sixty']
written_numbers['7'] = ['seven', 'seventy']
written_numbers['8'] = ['eight', 'eighty']
written_numbers['9'] = ['nine', 'ninety']

def convert_to_written(numb):
    # converts a number to written format
    numb = str(numb)
    if len(numb) == 1:
        return written_numbers[numb][0]
    elif len(numb) == 2 and numb[0] == '1':
        if numb[1] == '0':
            return 'ten'
        elif numb[1] == '1':
            return 'eleven'
        elif numb[1] == '2':
            return 'twelve'
        elif numb[1] == '3':
            return 'thirteen'
        elif numb[1] == '4':
            return 'fourteen'
        elif numb[1] == '5':
            return 'fifteen'
        elif numb[1] == '6':
            return 'sixteen'
        elif numb[1] == '7':
            return 'seventeen'
        elif numb[1] == '8':
            return 'eighteen'
        elif numb[1] == '9':
            return 'nineteen'
    elif len(numb) == 2:
        first = written_numbers[numb[0]][1]
        second = '-'
        if numb[1] != '0':
            third = written_numbers[numb[1]][0]
        else:
            return written_numbers[numb[0]][1]
        return first + second + third
    elif len(numb) == 4:
        return 'one thousand'
    else:
        first = written_numbers[numb[0]][0]
        if numb[1] == '0' and numb[2] == '0':
            second = ' hundred'
            return first + second
        else:
            second = ' hundred and '
            if numb[1] == '0':
                third = written_numbers[numb[2]][0]
                return first + second + third
            elif numb[1] == '1':
                return first + second + convert_to_written(numb[1:])
            else:
                third = written_numbers[numb[1]][1]
                if numb[2] != '0':
                    fourth = written_numbers[numb[2]][0]
                    return first + second + third + fourth
                else:
                    return first + second + third

def find_total_letters():
    count = 0
    for number in range(1, 1001):
        written = convert_to_written(number)
        count += len(re.findall('\w', written)) 
    return count

start_time = time.time()
print(find_total_letters())
print(time.time() - start_time)
```
